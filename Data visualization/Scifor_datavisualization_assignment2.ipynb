```python
import matplotlib.pyplot as plt

# Data
country = ['USA', 'China', 'India','Japan','Srilanka']
pop = [3100,5000, 7000, 2000, 1700]

plt.pie(pop, labels=country, startangle=180, colors=['blue', 'red', 'yellow','green','orange'])


plt.title('Population in Different Countries')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
cars = ['Ferrari', 'Lexus', 'BMW', 'Audi']
customers = [250, 40, 130, 75]


years = [2018, 2019, 2020, 2021, 2022]
ferrari = [200, 220, 240, 260, 280]
lexus = [35, 38, 42, 45, 48]
bmw = [120, 125, 130, 135, 140]
audi = [70, 72, 75, 78, 80]

# Create stacked bar chart
plt.figure(figsize=(10, 6))

plt.bar(years, ferrari, label='Ferrari', color='red')
plt.bar(years, lexus, label='Lexus', color='blue')
plt.bar(years, bmw, label='BMW', color='green')
plt.bar(years, audi, label='Audi', color='orange')


plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.title('Stacked Bar Chart of Customers for Different Car Brands Over Years')
plt.legend() 
plt.grid(True)

plt.show()
```
