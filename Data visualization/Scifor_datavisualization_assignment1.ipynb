```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Country': ['India', 'Germany', 'Japan', 'Italy', 'France'],
    'Low Price': [25000, 30000, 20000, 28000, 27000],
    'High Price': [40000, 45000, 35000, 42000, 38000]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.set_index('Country'), markers=True)
plt.title('Car Price Range in Different Countries')
plt.xlabel('Country')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Country': ['USA', 'Germany', 'Japan', 'Italy', 'France'],
    'Low Price': [25000, 30000, 20000, 28000, 27000],
    'High Price': [40000, 45000, 35000, 42000, 38000]
}

df = pd.DataFrame(data)

pal = sns.color_palette("pastel")

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.set_index('Country'), palette=pal, markers=True)
plt.title('Car Price Range in Different Countries')
plt.xlabel('Country')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'State': ['Maharashtra', 'Tamil Nadu', 'Karnataka', 'Uttar Pradesh', 'Gujarat'],
    '18-25': [12000, 10000, 11000, 13000, 9000],
    '26-30': [8000, 7000, 7500, 8500, 6000],
    '31-40': [6000, 5000, 5500, 6500, 4500],
    '41-50': [4000, 3000, 3500, 4500, 2500]
}

df = pd.DataFrame(data)

df_m = df.melt(id_vars='State', var_name='Age Group', value_name='Graduates')

plt.figure(figsize=(12, 8))
sns.barplot(data=df_m, x='State', y='Graduates', hue='Age Group')
plt.title('Number of Graduates in India')
plt.xlabel('State')
plt.ylabel('Number of Graduates')
plt.legend(title='Age Group')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'State': ['Maharashtra', 'Tamil Nadu', 'Karnataka', 'Uttar Pradesh', 'Gujarat'],
    '18-25': [12000, 10000, 11000, 13000, 9000],
    '26-30': [8000, 7000, 7500, 8500, 6000],
    '31-40': [6000, 5000, 5500, 6500, 4500],
    '41-50': [4000, 3000, 3500, 4500, 2500]
}
df = pd.DataFrame(data)

sns.violinplot(data=df.set_index('State'), palette='pastel')
plt.title('Number of Graduates in India')
plt.xlabel('Age Group')
plt.ylabel('Number of Graduates')
plt.xticks(rotation=45)
plt.legend(title='Age Group')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', hue='sex', kde=True, multiple='stack', palette='pastel')
plt.title('Dist Total Bill by Sex and Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='sex', y='total_bill', hue='smoker', palette='pastel', split=True)
plt.title('Total Bill by Sex and Smoker')
plt.xlabel('Sex')
plt.ylabel('Total Bill')
plt.legend(title='Smoker')
plt.show()
```
